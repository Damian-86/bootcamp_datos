[1] Consulta la API pública JSONPlaceholder utilizando el método GET para obtener:<br>

Lista de publicaciones (/posts)<br>
Lista de usuarios (/users)<br>
Lista de tareas (/todos)<br>

```python
# pip install requests
```

```python
import requests

# POSTS
url_posts = "https://jsonplaceholder.typicode.com/posts"

# Dame los post del servidor
response_posts = requests.get(url_posts)

# Convierto los datos en lista python
posts = response_posts.json()


# USERS
url_users = "https://jsonplaceholder.typicode.com/users"

response_users = requests.get(url_users)

users = response_users.json()


# TODOS
url_todos = "https://jsonplaceholder.typicode.com/todos"

response_todos = requests.get(url_todos)

todos = response_todos.json()

print(posts[:5])
print(users[:5])
print(todos[:5])
```

[4] Realiza una petición POST para crear una nueva publicación ficticia. Incluye el título, el cuerpo del mensaje y un userId.<br>

Muestra la respuesta JSON.<br>
Muestra el código de estado.<br>

```python
# Creamos datos del nuevo post
new_post = {
    "title": "Sprint 12 Practice",
    "body": "Esta es una publicación ficticia creada desde Python por Damian",
    "userId": 1
}

# Hacemos el post
response_post = requests.post(
    "https://jsonplaceholder.typicode.com/posts",
    json=new_post
)

print(response_post.json())
print("Status Code:", response_post.status_code)
```

[5] Realiza una petición PATCH para modificar parcialmente una publicación existente.<br>

Muestra la respuesta JSON.<br>
Muestra el código de estado.<br>

```python
# Creo lo que cambiaré (el titulo)
update_post = {
"title": "Titulo actualizado Sprint 12"
}

# Modifico el post/1 con patch 
response_patch = requests.patch("https://jsonplaceholder.typicode.com/posts/1", json=update_post)

print("\nPATCH - Publicación modificada")
print("Respuesta JSON:")
print(response_patch.json())
print("Status Code:", response_patch.status_code)
```

1. https://api.exchangerate.host/live

Obtiene tasas de cambio actuales.

2. https://api.exchangerate.host/convert

Filtros o parámetros opcionales

- symbols - seleccionar monedas específicas.

- end_date - fecha final (timeseries).

Ejemplo:

https://api.exchangerate.host/latest?base=EUR&symbols=USD,GBP

https://api.exchangerate.host/timeseries?start_date=2024-01-01&end_date=2024-01-05&base=EUR&symbols=USD

Formato respuesta

La API devuelve los datos en formato JSON.

```python
import requests

# Entrar con apikey
api_key = "991c466e9a2d4288eba629f161e52ee5"

# Endpoint /live
url = f"https://api.exchangerate.host/live?access_key={api_key}"

# Petición GET
response = requests.get(url)

print("Status Code:", response.status_code)

# Convertir respuesta a JSON
data = response.json()

print(data)

# Mostrar algunos campos del JSON
print("\nMoneda base:", data["source"])

print("USD:", data["quotes"]["USDEUR"])

print("GBP:", data["quotes"]["USDGBP"])
```

<h3 style="color:#5C7FA3; text-align:left; font-size:16px;">

[1] Utiliza la API de Open Data BCN para:<br>

Buscar un dataset de tu interés mediante package_search y package_show. .<br>

```python
import requests

url_opendata = "https://opendata-ajuntament.barcelona.cat/data/api/3/action"

params = {
    "q": "barris", 
    "rows": 5   
}

response = requests.get(f"{url_opendata}/package_search", params=params) 
response.raise_for_status() 

data = response.json() 

datasets = data["result"]["results"] 

print(f"Datasets encontrados: {len(datasets)}") 

# Bucle para decidir mejor que dataset usar
for i, d in enumerate(datasets):
    print(f"{i+1}. {d['title']}")
    print(f"   ID: {d['id']}")
    print(f"   Nº recursos: {len(d['resources'])}")
```

```python
# Elijo dataset 
dataset_id = datasets[1]["id"]

response = requests.get(f"{url_opendata}/package_show", params={"id": dataset_id})
response.raise_for_status()

dataset_detail = response.json()["result"] 

print("Título del dataset:") 
print(dataset_detail["title"])

print("\nRecursos disponibles:")
for r in dataset_detail["resources"]:
    print(f"- Nombre: {r['name']}")
    print(f"  Formato: {r['format']}")
    print(f"  Resource ID: {r['id']}")
```

<h3 style="color:#5C7FA3; text-align:left; font-size:16px;">

[2] De los resultados obtenidos, selecciona uno que tenga recursos disponibles en CSV o JSON.<br>

```python
datasets = data["result"]["results"]

# Buscar datasets que tengan CSV o JSON
filtered = [

    (d["title"], r["name"], r["format"], r["id"])

    for d in datasets

    for r in d["resources"]

    if r["format"].upper() in ["CSV", "JSON"]
]

# Mostrar solo los primeros 5
for dataset, resource, formato, rid in filtered[:5]:

    print("Dataset:", dataset)
    print("Resource:", resource)
    print("Formato:", formato)
    print("Resource ID:", rid)
    print("-"*40)
```
