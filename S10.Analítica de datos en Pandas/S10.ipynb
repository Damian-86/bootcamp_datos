[16] Crea una nueva columna que sea la fecha de nacimiento en formato Datetime a partir de las columnas día, mes y año. Utilizando esta columna crea una función que dada una fecha, te calcule la edad actual a día de hoy.<br>

```python
from datetime import datetime

# 16)  Crear columna de fecha de nacimiento 
df_limpio["Data_naixement"] = pd.to_datetime({"day": df_limpio["Dia"], "month": df_limpio["Mes"], "year": df_limpio["Any"]})

# Eliminamos columnas sobrantes
df_limpio = df_limpio.drop(columns=["Dia", "Mes", "Any"])

# Función para calcular edad actual
def calcular_edad(fecha_nacimiento):
    hoy = datetime.today()
    edad = hoy.year - fecha_nacimiento.year
    
    # Ajustar si aún no ha cumplido años este año
    if (hoy.month, hoy.day) < (fecha_nacimiento.month, fecha_nacimiento.day):
        edad -= 1 # si el mes y el dia de hoy es menor al dia y mes del cumpleaños restamos un año 
        
    return edad

# 17) Crear columna Edad_actual usando la función
df_limpio["Edat_actual"] = df_limpio["Data_naixement"].apply(calcular_edad)

df_limpio
```

```python
# Crear DataFrame de incrementos
df_incremento = pd.DataFrame({
    "Grup Professional": ["Grup A", "Grup B", "Grup C", "Grup D"],
    "Incremento": ["5%", "3,5%", "2%", "8%"]
})

# Convertir porcentaje a decimal
df_incremento["Incremento"] = (
    df_incremento["Incremento"]
    .str.replace("%", "", regex=False) # quita el simbolo %
    .str.replace(",", ".", regex=False) # cambia la coma a punto ya que python usa punto
    .astype(float) / 100  # convierte de str a float y lo pasa a porcentaje al dividir por 100
)

# 1) Unir con el DataFrame principal
df_limpio = df_limpio.reset_index().merge(
    df_incremento,
    on="Grup Professional",
    how="left"
).set_index("DNI")

# 2) Actualizar salario
df_limpio["Salari mensual"] = (
    df_limpio["Salari mensual"] +
    df_limpio["Salari mensual"] * df_limpio["Incremento"]
)

# Redondear
df_limpio["Salari mensual"] = df_limpio["Salari mensual"].round(2)

df_limpio
```

Nos interesa visitar todas las principales ciudades de España recorriendo la mínima distancia posible.<br>

Haz una función que dada la matriz de distancias y la ciudad de origen, haga una propuesta de ruta que sea lo más corta posible que puedas, devolviendo una lista con el orden de visita. <br>
Da también la distancia total recorrida.<br>
(EXTRA) Desde qué ciudad la ruta sería más corta con el algoritmo planteado<br>

```python
# Cargar matriz de distancias
df_matriz = pd.read_excel("matriu_distancies.xlsx", index_col=0)

# Eliminar ciudades que no podemos visitar en coche
df_matriz = df_matriz.drop(["Las Palmas de Gran Canaria", "Palma"], axis=0) # fila
df_matriz = df_matriz.drop(["Las Palmas de Gran Canaria", "Palma"], axis=1) # columna

display(df_matriz)
```

```python
def ruta_mas_corta(df, origen):

    ciudad_actual = origen
    visitadas = [origen]
    distancia_total = 0

    while len(visitadas) < len(df): # mientras no hayamos visitado todas las ciudades

        # Distancias desde la ciudad actual
        distancias = df.loc[ciudad_actual] # con loc busca la fila de esa ciudad y lo guarda en una serie llamada distancias

        # Quita las ciudades ya visitadas de la variable distancias
        distancias = distancias.drop(visitadas) 

        # Elegir la más cercana
        siguiente = distancias.idxmin() # idxmin() devuelve el índice del valor mínimo, osea me devuelve solo el nombre de la ciudad con el menor kilometraje

        # Sumar distancia recorrida
        distancia_total += distancias[siguiente]

        # Añadir ciudad a la ruta
        visitadas.append(siguiente)

        # Actualizar ciudad actual
        ciudad_actual = siguiente

    return visitadas, distancia_total

# Ejecución:
ruta, distancia = ruta_mas_corta(df_matriz, "Valencia")

print("Ruta:", ruta)
print("Distancia total:", distancia)
```

```python
# (EXTRA) mejor ruta de origen
def mejor_origen(df):

    mejor_distancia = float("inf") # infinito, lo usamos para que cualquier distancia sea menor al principio
    mejor_ciudad = None # aquí guardaremos la ciudad con mejor resultado, no hago una lista por que espero un solo resultado

    for ciudad in df.index: # matriz.index contiene todas las ciudades

        ruta, distancia = ruta_mas_corta(df, ciudad) # llamamos a la función anterior

        if distancia < mejor_distancia:  # si encontramos una ruta mejor:
            mejor_distancia = distancia
            mejor_ciudad = ciudad
            mejor_ruta = ruta

    return mejor_ciudad, mejor_distancia, mejor_ruta # devolver mejor ciudad, 

# Ejecución:
ciudad, distancia, ruta = mejor_origen(df_matriz)

print("Mejor ciudad de origen:", ciudad)
print("Distancia mínima:", distancia)
print("Mejor Ruta:", ruta)
```
