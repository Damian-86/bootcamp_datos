<h1 style="color:#3A6EA5; text-align:left;">
<b>Calculadora de Índice de Masa Corporal</b>
</h1>

<h3 style="color:#5C7FA3; text-align:left;">
Programa para calcular y clasificar el IMC según la OMS
</h3>

</div>

```python
def calculador_imc():
    try:
        # Solicitar datos al usuario
        peso = float(input("Introduce tu peso en kg: "))
        altura = float(input("Introduce tu altura en metros (ejemplo 1.75): "))

        # Validaciones 
        if peso <= 0 or altura <= 0:
            print("El peso y la altura deben ser mayores que cero.")
            return
        if peso < 30 or peso > 300:
            print("El peso introducido no parece razonable.")
            return
        if altura < 1.0 or altura > 2.5:
            print("La altura introducida no parece razonable.")
            return

        # Cálculo del IMC
        imc = peso / (altura ** 2)

        # Clasificamos por grado de peso
        if imc < 18.5:
            categoria = "Bajo peso"
        elif imc < 25:
            categoria = "Peso normal"
        elif imc < 30:
            categoria = "Sobrepeso"
        elif imc < 35:
            categoria = "Obesidad grado I"
        elif imc < 40:
            categoria = "Obesidad grado II"
        else:
            categoria = "Obesidad grado III"

        # Resultado
        print(f"Tu IMC es: {imc:.2f}") #.2f son dos decimales (2 float)
        print(f"Clasificación: {categoria}")

    except ValueError: # con try intenta ejecutar el codigo, si ocurre un value error imprime este texto
        print("Debes introducir valores numéricos válidos.")

# Ejecución
calculador_imc()
```

<h1 style="color:#3A6EA5; text-align:left;">
<b>Convertidor de Temperaturas</b>
</h1>

<h3 style="color:#5C7FA3; text-align:left;">
Conversión automática entre Celsius, Fahrenheit, Kelvin, Rankine y Réaumur <br>
El programa convierte cualquier temperatura primero a Celsius y después usa ese valor para calcular todas las demás unidades. <br>
Los resultados se almacenan en un diccionario para poder guardarlos y utilizarlos posteriormente.
</h3>
</div>

```python
def convertir_temperatura(valor, unidad):

    # Validación
    if not isinstance(valor, (int, float)):
        return "Error: la temperatura debe ser numérica"
    if not isinstance(unidad, str):
        return "Error: la unidad debe ser texto"

    unidad = unidad.upper()

    # Conversión a Celsius usando diccionario y lambda
    a_celsius = {
        "C": lambda x: x,
        "F": lambda x: (x - 32) * 5/9,
        "K": lambda x: x - 273.15,
        "RA": lambda x: (x - 491.67) * 5/9,
        "RE": lambda x: x * 5/4
    }

    if unidad not in a_celsius:
        return "Unidad no válida, usa C, F, K, RA o RE."

    celsius = a_celsius[unidad](valor)

    # Convertir desde Celsius a todas las unidades
    conversiones = {
        "Celsius": round(celsius, 2),                         
        "Fahrenheit": round((celsius * 9/5) + 32, 2),
        "Kelvin": round(celsius + 273.15, 2),                   
        "Rankine": round((celsius + 273.15) * 9/5, 2),
        "Réaumur": round(celsius * 4/5, 2)
    }

    return conversiones

# Ejecucion
resultado = convertir_temperatura(40, "k")
print(resultado)
```

<h1 style="color:#3A6EA5; text-align:left;">
<b>Contador de Palabras en un Texto</b>
</h1>

<h3 style="color:#5C7FA3; text-align:left;">
Conteo de palabras y cálculo de longitud media
</h3>

</div>

```python
import string

def analizar_texto(texto):
    
    # Validación
    if not isinstance(texto, str):
        return "El texto debe ser una cadena de caracteres."
    if texto.strip() == "":  # elimina los espacios al principio y la final
        return "El texto no puede estar vacío."
    
    # Pasar todo a minúsculas
    texto = texto.lower()
    
    # Eliminar signos de puntuación, interrogación y exclamación
    texto = texto.translate(str.maketrans("", "", string.punctuation + "¿¡"))   
    
    # Separar en palabras
    palabras = texto.split()
    
    # Contar palabras
    conteo = {}
    
    for palabra in palabras:
        if palabra in conteo:
            conteo[palabra] += 1
        else:
            conteo[palabra] = 1
    
    # Calcular longitud media 
    total_letras = sum(len(palabra) for palabra in palabras)
    longitud_media = round(total_letras / len(palabras), 2)
    
    # Devolver resultados en diccionario
    resultado = {
        "conteo_palabras": conteo,
        "longitud_media": longitud_media
    }
    
    return resultado

# Ejecución
resultado = analizar_texto("¿Son las cinco menos cinco, faltan cinco para las cinco, cuántas veces dije cinco?")
print(resultado)
```

<h1 style="color:#3A6EA5; text-align:left;">
<b>Diccionario Inverso</b>
</h1>

<h3 style="color:#5C7FA3; text-align:left;">
Intercambio de claves y valores con gestión de valores duplicados
</h3>

</div>

```python
def reverse_dictionary(diccionario):
    
    # Validación
    if not isinstance(diccionario, dict):
        return "El argumento debe ser un diccionario."
    
    diccionario_inverso = {}
    duplicados = False  # Esta variable sirve para saber si dos claves tienen el mismo valor
    
    for clave, valor in diccionario.items():
        
        if valor not in diccionario_inverso:
            diccionario_inverso[valor] = clave
        else:
            # Si ya existe, hay duplicado
            duplicados = True
            
            # Si no es lista, lo convierte a lista
            if not isinstance(diccionario_inverso[valor], list):
                diccionario_inverso[valor] = [diccionario_inverso[valor]]
            
            diccionario_inverso[valor].append(clave)
    
    if duplicados:
        print("Existen valores duplicados:")
    
    return diccionario_inverso

# Ejecución
print(reverse_dictionary({'x':'celeste', 'y':'blanco', 'z':'celeste'}))
```

<h1 style="color:#3A6EA5; text-align:left;">
<b>Contador y Ordenador de Palabras</b>
</h1>

<h3 style="color:#5C7FA3; text-align:left;">
Frecuencia y agrupación alfabética en archivos txt<br>
La función contador_ordenador lee un archivo de texto, limpia el contenido eliminando puntuación y pasando todo a minúsculas, <br>
cuenta cuántas veces aparece cada palabra y organiza los resultados en un diccionario agrupado por la letra inicial de cada palabra.
</h3>

</div>

```python
def contador_ordenador(ruta_archivo):
    
    # Validación
    if not isinstance(ruta_archivo, str):
        return "La ruta debe ser texto."
    
    try:
        with open(ruta_archivo, "r", encoding="utf-8") as archivo:  # "r" significa modo lectura y "utf-8" permite leer correctamente tildes y caracteres especiales
            texto = archivo.read()
    except FileNotFoundError:  # Si el archivo no existe, el programa no se rompe y devuelve ese mensaje
        return "Archivo no encontrado."
    
    # Pasar a minúsculas
    texto = texto.lower()
    
    # Eliminar puntuación, signos de pregunta y exclamación
    texto = texto.translate(str.maketrans("", "", string.punctuation + "¿¡")) # maketrans() Crea una tabla que indica qué caracteres deben eliminarse y translate() aplica esa tabla y borra esos caracteres del texto.
    
    palabras = texto.split()
    
    # Contar palabras
    conteo = {}
    
    for palabra in palabras:
        if palabra in conteo:
            conteo[palabra] += 1
        else:
            conteo[palabra] = 1
    
    # Ordenar palabras alfabéticamente
    palabras_ordenadas = sorted(conteo.items())
    
    # Agrupar por letra inicial
    resultado = {}
    
    for palabra, frecuencia in palabras_ordenadas:
        letra = palabra[0]
        
        if letra not in resultado:
            resultado[letra] = {}
        
        resultado[letra][palabra] = frecuencia
    
    return resultado

# Ejecución
resultado = contador_ordenador("tu_me_quieres_blanca.txt")
from pprint import pprint
pprint(resultado, width=200) # Usé pprint visualizar mejor las key una abajo de la otra.
```

<h1 style="color:#3A6EA5; text-align:left;">
<b> Conversión de Tipos de Datos</b>
</h1>

<h3 style="color:#5C7FA3; text-align:left;">
Separación automática de valores numéricos y no numéricos
</h3>

</div>

```python
def conversion(lista):
    
    if not isinstance(lista, list):
        return "El argumento debe ser una lista."
    
    convertidos = []
    no_convertidos = []
    
    for elemento in lista:
        
        try:
            # Si el elemento dentro de la lista es una lista, recorrer esta sublista
            if isinstance(elemento, list):
                for sub in elemento:
                    try:
                        convertidos.append(float(sub))
                    except:
                        no_convertidos.append(sub)
            else:
                convertidos.append(float(elemento))
        
        except:
            no_convertidos.append(elemento)
    
    return convertidos, no_convertidos

# Ejecución
datos = [ '1.3', 'one', '1e10', 'seven', '3-1/2', ('2',1,1.4,'not-a-number'), [1,2,'3','3.4']]

resultado = conversion(datos)
pprint(resultado)
```

<h3 style="color:#5C7FA3; text-align:left;">
Creación automática de contraseñas seguras con parámetros configurables

```python
# pip install pyperclip
```

```python
import numpy as np

def crear_contraseña(longitud, mayusculas=True, minusculas=True, numeros=True, signos=False):
    
    # Validación de longitud que sea entero y mayor que 0
    if not isinstance(longitud, int) or longitud <= 0:
        return "La longitud debe ser un número entero positivo."
    
    # Construcción de caracteres
    caracteres_disponibles = ""
    obligatorios = []
    
    if mayusculas:
        caracteres_disponibles += string.ascii_uppercase  # agrega una mayuscula
        obligatorios.append(np.random.choice(list(string.ascii_uppercase))) # aqui asegura una mayuscula
    
    if minusculas:
        caracteres_disponibles += string.ascii_lowercase   # añade miníscula
        obligatorios.append(np.random.choice(list(string.ascii_lowercase))) # las asegura
    
    if numeros:
        caracteres_disponibles += string.digits  # añade números
        obligatorios.append(np.random.choice(list(string.digits))) # los asegura
    
    if signos:
        caracteres_especiales = ",.-_$?!"
        caracteres_disponibles += caracteres_especiales    # añade caracteres
        obligatorios.append(np.random.choice(list(caracteres_especiales))) # los asegura
    
    if caracteres_disponibles == "":    # verkifiva que haya caracteres disponibles
        return "Debe activarse al menos un tipo de carácter."
    
    if longitud < len(obligatorios):   # verifica que la longitud sea suficiente
        return "La longitud es demasiado corta para los criterios seleccionados." 
    
    # Generar resto aleatorio
    restantes = longitud - len(obligatorios) # calcula cuántos caracteres faltan
    contraseña = obligatorios + list(np.random.choice(list(caracteres_disponibles), restantes)) 
    
    # Mezclar para que no queden los obligatorios al inicio
    np.random.shuffle(contraseña)
    
    return "".join(contraseña) # une todos los caracteres de la lista sin poner nada entre ellos

# Ejecución
import pyperclip
password = crear_contraseña(12, True, True, True, True)
pyperclip.copy(password)
print("Password:", password)
print("Contraseña copiada al portapapeles.")
```

<h1 style="color:#3A6EA5; text-align:left;">
<b> Procesador Automático de Resultados Deportivos</b>
</h1>

<h3 style="color:#5C7FA3; text-align:left;">
Cálculo de estadísticas y clasificación a partir de registros históricos de partidos
</h3>

</div>

```python
def procesar_partidos(ruta):

    equipos = {}

    with open(ruta, "r", encoding="utf-8") as archivo:

        for linea in archivo:

            # Elimina espacios o saltos de línea al inicio o al final
            linea = linea.strip()

            # Separar columnas
            partes = linea.split("\t")

            local = partes[0]
            marcador = partes[1]
            visitante = partes[2]

            # Separar goles
            goles = marcador.split("-")
            goles_local = int(goles[0])  # 0 y 1 son las posiciones en la lista
            goles_visitante = int(goles[1])

            # Crear equipos si no existen
            if local not in equipos:
                equipos[local] = {
                    "goles_favor": 0,
                    "goles_contra": 0,
                    "puntos": 0
                }

            if visitante not in equipos:
                equipos[visitante] = {
                    "goles_favor": 0,
                    "goles_contra": 0,
                    "puntos": 0
                }

            # Sumar goles
            equipos[local]["goles_favor"] += goles_local
            equipos[local]["goles_contra"] += goles_visitante

            equipos[visitante]["goles_favor"] += goles_visitante
            equipos[visitante]["goles_contra"] += goles_local

            # Asignar puntos
            if goles_local > goles_visitante:
                equipos[local]["puntos"] += 3

            elif goles_local < goles_visitante:
                equipos[visitante]["puntos"] += 3

            else:
                equipos[local]["puntos"] += 1
                equipos[visitante]["puntos"] += 1


    # Equipo más goleador
    mas_goleador = max(equipos, key=lambda equipo: equipos[equipo]["goles_favor"]) # busca el valor max bajo el criterio key = lambda equipo basado en goles a favor

    # Equipo más goleado
    mas_goleado = max(equipos, key=lambda equipo: equipos[equipo]["goles_contra"]) # idem con max de goles en contra

    # Clasificación ordenada
    clasificacion = sorted(
        equipos.items(),
        key=lambda item: item[1]["puntos"], # ordena los equipos en funcion de los puntos (que estan en la posicion 1 del valor de la key "puntos")
        reverse=True # de mayor a menor
    )
    
    # Resultado final
    resultado = {
        "goles_por_equipo": sorted({equipo: equipos[equipo]["goles_favor"] for equipo in equipos}
                            .items(),key=lambda x: x[1],reverse=True), # Ordenados de mayor a menor por el segundo valor de la tupla (x[1]) que son los goles.
        "equipo_mas_goleador": mas_goleador,
        "equipo_mas_goleado": mas_goleado,
        "clasificacion": clasificacion
    }

    return resultado

# Ejecución
resultado = procesar_partidos("historic_partits.txt")
pprint(resultado, width=200)
```
